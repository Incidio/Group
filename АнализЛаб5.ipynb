<h3 align="left"> Подключены библиотеки </h3>

```r
D <- read.csv("BankChurners.csv")
D <- as.data.frame(D)

names(D)           

## Проверим тип данных 
str(D)
```

%%HTML
<h3 align="left"> Датасет подготовлен к работе </h3>


</h3>

```r
# Произвдена нормализаиця 
cols <- sapply(D, is.numeric)
D[cols] <- scale(D[cols])

# Для наилучшего предсказания избавимся от выбросов

outliers <- function(x) {

  Q1 <- quantile(x, probs=.25)
  Q3 <- quantile(x, probs=.75)
  iqr = Q3-Q1

 upper_limit = Q3 + (iqr*1.5)
 lower_limit = Q1 - (iqr*1.5)

 x > upper_limit | x < lower_limit
}

remove_outliers <- function(df, cols = names(df)) {
  for (col in cols) {
    df <- df[!outliers(df[[col]]),]
  }
  df
}

remove_outliers(D, c('Customer_Age', 'Months_on_book', 'Total_Relationship_Count', 'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal', 'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt',
                     'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio'))

# Избавимся от ненужных столбцов

D <- subset(D, select = -c(CLIENTNUM))
D <- subset(D, select = -c(Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1))
D <- subset(D, select = -c(Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2))

# Категориальные переменные приведем к количественным

D$Attrition_Flag= ifelse(D$Attrition_Flag =="Attrited Customer","1","0")
D$Attrition_Flag = as.factor(D$Attrition_Flag)

D$Gender <- as.integer(factor(D$Gender))

D$Education_Level <- as.integer(factor(D$Education_Level))
D$Marital_Status <- as.integer(factor(D$Marital_Status))
D$Income_Category <- as.integer(factor(D$Income_Category))
D$Card_Category <- as.integer(factor(D$Card_Category))


table(D$Attrition_Flag)

dim(D)
```

%%HTML
<h3 align="left"> Ансамбли моделей </h3>


</h3>

%%HTML
<h3 align="left"> Датасет разделен на тренировочную и тестовую выборки </h3>


</h3>

```r
set.seed(123)

sample <- sample(c(TRUE, FALSE), nrow(D), replace=TRUE, prob=c(0.7,0.3))
train <- D[sample, ]
test <- D[!sample, ]
```

%%HTML
<h3 align="left"> Бэггинг - рассматриваются однородные модели, которые обучаются независимо и параллельно, а затем их результаты просто усредняются. </h3>


</h3>

```r
bag <- bagging(
 formula = Attrition_Flag ~ .,
 data = train,
 nbagg = 10, 
 coob = TRUE,
 control = rpart.control(minsplit = 2, cp = 0)
)
bag

y_predbag <- predict(bag, newdata=test)
# Confusion Matrix
cm <- table(test$Attrition_Flag, y_predbag)

# Model Evaluation
confusionMatrix(cm)
```

%%HTML
<h3 align="left"> Random Forest - алгоритм заключающийся в использовании ансамбля решающих деревьев </h3>


</h3>

```r
randomF <- randomForest(
 formula = Attrition_Flag ~ .,
 data = train,
 mtry = 4,
 ntree = 501
)
randomF

y_predrand <- predict(randomF, newdata=test)
# Confusion Matrix
cm <- table(test$Attrition_Flag, y_predrand)

# Model Evaluation
confusionMatrix(cm)
```

%%HTML
<h3 align="left"> Бустинг — это метод, в котором строится множество базовых алгоритмов из одного семейства, объединяющихся затем в более сильную модель. </h3>


</h3>

```r
# Разделите данные на матрицу признаков и вектор меток
X_train <- as.matrix(train[, names(train) != "Attrition_Flag"])
y_train <- as.integer(as.character(train$Attrition_Flag))
# Преобразуйте в матрицу DMatrix для использования с XGBoost
dtrain <- xgb.DMatrix(data = X_train, label = y_train)
# Задайте параметры модели
params <- list(
  objective = "binary:logistic",  # Для бинарной классификации
  eval_metric = "logloss"         # Метрика оценки качества модели
)

# Обучите модель с заданными параметрами
model <- xgboost(params = params, data = dtrain, nrounds = 50)

# Для предсказаний на новых данных
X_newm <- as.matrix(test[, names(test) != "Attrition_Flag"])
X_new <- as.matrix(X_newm)  # Подготовьте новые данные

dtest <- xgb.DMatrix(data = X_new)  # Преобразуйте в DMatrix

predictions <- predict(model, dtest)  # Получите предсказания
```

```r
threshold <- 0.5

# Преобразование вероятностей в 0 и 1
binary_values <- ifelse(predictions > threshold, 1, 0)

y_test <- as.integer(as.character(test$Attrition_Flag))
# Confusion Matrix
cm <- table(y_test, binary_values)

# Model Evaluation
confusionMatrix(cm)
```
